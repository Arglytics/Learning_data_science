```python
import pandas as pd
```

```python
df = pd.read_csv("glassdoor_jobs.csv")
df.head(10)
```

```python
#salary parsing.
```

```python
#checks if per hour and employer provided exist by puting a 1 or 0 if they donot exist
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
df.head()
```

```python
#removing the -1
df = df[df['Salary Estimate'] != '-1']


```

```python
df.head()
```

```python
#removing the word glassdoor est. in the Salary Estimate
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
#salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary.head()
```

```python
#removing the k on the salaries and dollar sign.
minus_Kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
#minus_kd = salary.apply(lambda x: x.replace('K','').replace('$',''))
#print(minus_kd)(150
print(minus_kd[:150])
```

```python
#removing the hour from the column of salary estimator or from the column minus_kd and we have columns of per hour 
#and employer provided showing with one and zeros
min_hr = minus_Kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
#min_hr = minus_kd.apply(lambda x: x.lower().replace('per hour','').replace('employer provided',''))
min_hr.head()
```

```python
#getting the minimum salary by splitting off the hyphen...so we created the column min salary.
df['min_salary'] = min_hr.apply(lambda x: int(x.split('-')[0]))  #0 shows that we are taking the first element
df['max_salary'] = min_hr.apply(lambda x: int(x.split('-')[1])) #1 shows that we are taking the second element.
df['avg_salary'] = (df.min_salary+df.max_salary)/2

df.head()
```

```python
#company name text only.
#state field.
#age of company..
#parsing of jo description(python,etc.
```

```python
#company name text only.
df['company_txt'] = df.apply(lambda x: x['Company Name'] if x['Rating'] <0 else x['Company Name'][:-1], axis = 1)
```

```python
df.head()
```

```python
#state field.  what state is the job in
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
```

```python
df.head()
```

```python
#seein the number of jobs in each state.
df.job_state.value_counts()
```

```python

df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df.head()
```

```python
#age of company..
df['age'] = df.Founded.apply(lambda x: x if x <1 else 2020 - x)
df.head()
```

```python
#parsing of jo description(python,etc.
df['Job Description'][0]   #displaaying the first cell of the job description column
```

```python
#checking out the tools needed for a particular job like python,R,aws, excel
```

```python
#Python   checking out job descriptions with python
df['python_yn'] = df['Job Description'].apply(lambda x: 1 if 'python' in x.lower() else 0) #those with python give 1 or else 0
df.python_yn.value_counts()   #counting the jobs with thedescription python 
```

```python
#r studio
df['R_yn'] = df['Job Description'].apply(lambda x: 1 if 'r studio' in x.lower() or 'r studio' in x.lower() else 0) #those with python give 1 or else 0
df.R_yn.value_counts() 
```

```python
#spark
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'spark' in x.lower() else 0) #those with python give 1 or else 0
df.spark.value_counts() 
```

```python
#aws
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0) #those with python give 1 or else 0
df.aws.value_counts()
```

```python
#excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'excel' in x.lower() else 0) #those with python give 1 or else 0
df.excel.value_counts()
```
